```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('sinasc_RO_2019.csv')

df.head()
```

```python
df.columns
```

####  1) Na base do SINASC do DataSUS, faça um boxplot da idade da mãe.

```python
df[['IDADEMAE']].boxplot()
plt.show()
```

####  2) Faça um boxplot do peso do bebê

```python
df[['PESO']].boxplot()
plt.show()
```

####  3) Faça um histograma da idade da mãe

```python
idade_filtrada = df[df['IDADEMAE'].notnull()]

sns.displot(data=idade_filtrada, 
            x='IDADEMAE', 
            alpha=0.25, 
            bins=20, 
            element='step', 
            kde=True)

plt.show()
```

#### 4) Faça um histograma do peso do bebê

```python
peso_filtrado = df[df['PESO'].notnull()]

sns.displot(data=peso_filtrado, 
            x='PESO', 
            alpha=0.25, 
            bins=20, 
            element='step', 
            kde=True)

plt.show()
```
