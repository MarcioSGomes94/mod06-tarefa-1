#### 1) A base 'sinasc_RO_2019.csv' contém dados do sistema de nascidos vivos do DataSUS, cujo dicionário de dados está no arquivo 'Estrutura_SINASC_para_CD.pdf'. Carregue esses arquivos.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('sinasc_RO_2019.csv')

df.head()
```

```python
df.columns
```

####  2) Monte um gráficos de barras com as variáveis APGAR1, APGAR5, SEXO, GESTACAO, ESTCIVMAE.

```python
fig, axes = plt.subplots(2, 3, figsize=(15, 10))


sns.countplot(ax=axes[0, 0], x='APGAR1', data=df)
axes[0, 0].set_title('Contagem de APGAR1')
axes[0, 0].set_xlabel('APGAR1')
axes[0, 0].set_ylabel('Frequência')
axes[0, 0].tick_params(axis='x', rotation=45)

sns.countplot(ax=axes[0, 1], x='APGAR5', data=df)
axes[0, 1].set_title('Contagem de APGAR5')
axes[0, 1].set_xlabel('APGAR5')
axes[0, 1].set_ylabel('Frequência')
axes[0, 1].tick_params(axis='x', rotation=45)

sns.countplot(ax=axes[1, 0], x='SEXO', data=df)
axes[1, 0].set_title('Contagem de SEXO')
axes[1, 0].set_xlabel('SEXO')
axes[1, 0].set_ylabel('Frequência')
axes[1, 0].tick_params(axis='x', rotation=45)

sns.countplot(ax=axes[1, 1], x='GESTACAO', data=df)
axes[1, 1].set_title('Contagem de GESTACAO')
axes[1, 1].set_xlabel('GESTACAO')
axes[1, 1].set_ylabel('Frequência')
axes[1, 1].tick_params(axis='x', rotation=45)

sns.countplot(ax=axes[1, 2], x='ESTCIVMAE', data=df)
axes[1, 2].set_title('Contagem de ESTCIVMAE')
axes[1, 2].set_xlabel('ESTCIVMAE')
axes[1, 2].set_ylabel('Frequência')
axes[1, 2].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```

#### 3) Monte um gráfico de barras mostrando o APGAR5 médio por CONSULTAS

```python

df[['CONSULTAS']].head()

```

```python
apgar5_media = df.groupby('CONSULTAS')['APGAR5'].mean().reset_index()

plt.figure(figsize=(10, 6))
sns.barplot(x='CONSULTAS', y='APGAR5', data=apgar5_media)

plt.title('APGAR5 Médio por CONSULTAS')
plt.xlabel('CONSULTAS')
plt.ylabel('APGAR5 Médio')
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```

####  4) Monte um painel com 4 gráficos mostrando as frequencias de SEXO, GESTACAO, ESTCIVMAE e CONSULTAS

```python
fig, axes = plt.subplots(2, 2, figsize=(15, 10))


sns.countplot(ax=axes[0, 0], x='SEXO', data=df)
axes[0, 0].set_title('Contagem de SEXO')
axes[0, 0].set_xlabel('SEXO')
axes[0, 0].set_ylabel('Frequência')
axes[0, 0].tick_params(axis='x', rotation=45)

sns.countplot(ax=axes[0, 1], x='GESTACAO', data=df)
axes[0, 1].set_title('Contagem de GESTACAO')
axes[0, 1].set_xlabel('GESTACAO')
axes[0, 1].set_ylabel('Frequência')
axes[0, 1].tick_params(axis='x', rotation=45)

sns.countplot(ax=axes[1, 0], x='ESTCIVMAE', data=df)
axes[1, 0].set_title('Contagem de ESTCIVMAE')
axes[1, 0].set_xlabel('ESTCIVMAE')
axes[1, 0].set_ylabel('Frequência')
axes[1, 0].tick_params(axis='x', rotation=45)

sns.countplot(ax=axes[1, 1], x='CONSULTAS', data=df)
axes[1, 1].set_title('Contagem de CONSULTAS')
axes[1, 1].set_xlabel('CONSULTAS')
axes[1, 1].set_ylabel('Frequência')
axes[1, 1].tick_params(axis='x', rotation=45)


plt.tight_layout()
```
